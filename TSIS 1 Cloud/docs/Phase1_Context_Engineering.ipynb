### **ANALYSIS FRAMEWORK (THE CAGAN TEST)**
1. MERCENARIES vs. MISSIONARIES: Analyze if teams are empowered to solve customer problems or are simply digitizing traditional banking requirements.
2. PROJECT vs. PRODUCT FUNDING: Look for evidence of 3-year "Strategic Cycles" vs. persistent, outcome-based funding.
3. IT vs. PRODUCT: Determine if Technology is a back-office cost center or the core of the business value proposition.
4. OUTPUT vs. OUTCOME: Does the bank measure success by "features launched" or "customer value/engagement"?
