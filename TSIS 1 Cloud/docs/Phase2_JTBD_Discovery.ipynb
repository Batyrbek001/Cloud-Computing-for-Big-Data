# **Phase 2 — JTBD Discovery**

### **Role: Product Manager / JTBD Lead**
**Company:** Halyk Bank (HSBK)

**Domain:** Corporate Legal Compliance & KYC (Know Your Customer)

**2.1 Purpose of Phase 2**

The purpose of Phase 2 is to drill down into a specific operational bottleneck identified during the Halyk Bank Strategic Diagnosis. While Halyk Bank has successfully digitized its retail front-end, the internal Corporate Legal Compliance functions remain "Digital Immigrants," relying on heavy manual oversight.

This phase utilizes the Jobs to be Done (JTBD) framework and the 5 Whys technique to move past the symptom of "slowness" and uncover the root psychological and systemic blockers. The goal is to design an AI Agent that doesn't just "process data," but provides the psychological safety required for faster decision-making in a high-stakes banking environment.

**2.2 Identified Problem (Derived from Phase 1 Insights)**

Phase 1 revealed that Halyk Bank operates with a "multi-level governance system" (ESG Report, p. 6) and high centralized control. This translates to an internal culture where:

* **Complex Corporate Onboarding (SME/B2B)** is slow due to manual Charter/Statute (Устав) reviews.

* **Risk Aversion** is high because of strict National Bank of Kazakhstan (NBK) regulations.

* **"Digital Gap":** Retail is 90% automated, but complex corporate legal review is still 90% manual.

**2.3 Simulated User Persona (AI Persona)**

**AI Persona Profile**

* **Role:** Senior Legal Compliance Officer

* **Company:** Halyk Bank

* **Experience:** 10+ years in Banking Law and AML (Anti-Money Laundering)

* **Context:** Responsible for verifying complex corporate structures (SME and Large Corporates) before account opening.

* **Primary Pain:** The cognitive exhaustion of "hunting" for ultimate beneficial owners (UBOs) and restrictive clauses across 100+ page PDF documents.

* **Motivation:** Ensuring 100% compliance with NBK and Basel III standards to avoid personal sanctions.

**2.4 JTBD Interview Log (5 Whys)**

| Step | Product Manager Question | AI Persona Response |
| :--- | :--- | :--- |
| **1** | Why does it take 48+ hours to clear a single corporate client’s legal documents? | Because I have to read every page of their Articles of Association and Board Resolutions to map out the ownership structure and authorities. |
| **2** | Why can’t we use simple keyword search to speed this up? | Keywords don't tell the whole story. I need to understand the intent of a clause—for example, if a CEO has "limited" authority that could lead to an unauthorized loan. |
| **3** | Why is it necessary for a Senior Officer like you to do the initial reading? | Because if a junior or a basic OCR tool misses a "negative pledge" or a hidden sanction risk, my signature is on the final approval. I don't trust the current tools. |
| **4** | Why is trust in automation so low in your department? | Because the "system" just gives me a "Pass/Fail" or a huge list of highlights. It doesn't tell me **why** something is a risk or where exactly the bank’s policy is being violated. |
| **5** | Why is "the why" more important than the speed of the tool? | Because my job isn't to read; my job is to **defend my decision** to the auditors. If I can't explain why I missed a risk, I lose my license. I need a "second pair of eyes" that thinks like a lawyer, not just a scanner. |


**2.5 Key Insight (Root Cause)**

The core problem is Explainability Fatigue. The Legal Compliance team doesn't need "faster OCR"; they need traceable evidence. The current bottleneck is caused by the time spent cross-referencing document text with Halyk Bank’s internal Risk Policy. The officers are not afraid of work; they are afraid of indefensible oversight.

**2.6 Job to Be Done (JTBD Statement)**

When I am performing a legal compliance review for a new corporate client, I want an AI that highlights specific risks and maps them directly to our internal bank policies with text citations, so that I can confidently sign off on the onboarding without the fear of an indefensible audit finding.

**2.7 Product Requirements Document (PRD): "Halyk LegalEye"**

**Problem Statement**

Corporate onboarding at Halyk Bank is delayed by manual legal reviews. Compliance officers spend 80% of their time searching for information and only 20% analyzing it, leading to burnout and high operational costs.

**Target User**

* Halyk Bank Legal Compliance Officers

* AML/KYC Specialists

* Internal Auditors (for post-approval checks)

**Functional Requirements**

1. **Semantic Clause Extraction:** Move beyond keyword search to identify "Authorities," "Ownership Limits," and "Sanction Links" using Natural Language Processing.

2. **Policy Mapping:** Every flagged clause must be accompanied by a side-by-side reference to the specific Halyk Internal Policy or NBK Regulation it violates.

3. **Audit Trail Generation:** The AI must produce a "Compliance Memo" draft that cites page and line numbers, which the officer can use as their defense file.

4. **"Confidence Score" with Reasoning:** If the AI is unsure, it must state: "I found this clause ambiguous because [Reasoning]."

**Non-Goals**

* Automatically "Rejecting" or "Approving" a client (Decision remains human).

* Storing sensitive client documents in a public cloud.

* Direct editing of the client's legal documents.

**Constraints & Governance**

* **Data Residency:** All processing must occur on Halyk Bank’s private cloud (local Kazakhstan servers).

* **Explainability:** No "black box" logic; every AI suggestion must be traceable to the source PDF.

* **Human-in-the-Loop:** The AI Agent acts as an Assistant (Copilot), not an Autopilot.

**2.8 Phase 2 Output Summary**

This discovery shows that the value of an AI Agent for Halyk Bank is not in "replacing lawyers," but in augmenting their confidence.

* **ROI Impact:** Moving from 48 hours to 4 hours per complex client review.

* **Risk Impact:** Reducing "Human Error" by providing a systematic second check against the 2025-2027 Strategy goals.
