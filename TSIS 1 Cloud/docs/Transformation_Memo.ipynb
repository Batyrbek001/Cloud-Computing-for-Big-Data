# **Transformation Memo: Board of Directors, Halyk Bank**

**TO:** Board of Directors, Halyk Bank (HSBK)

**FROM:** Product Operating Model Consultant

**DATE:** February 9, 2026

**SUBJECT:** Urgent Pivot Required: From Digital Immigrant to Product-Led Leader

### **1. Executive Summary**

Halyk Bank is currently a "Digital Immigrant" operating as a high-performing Feature Factory. While the bank has achieved massive scale through the digitization of traditional banking services, it remains trapped in an "IT Service" mindset. The current structure relies on top-down, stakeholder-driven roadmaps that treat technology as a cost center rather than the core business driver. To survive the next wave of fintech disruption, Halyk must move away from "mercenary" delivery teams toward empowered missionary teams that own customer outcomes.

### **2. Evidence of Feature Factory Behavior**

Internal signals from official disclosures reveal an output-centric culture that prioritizes "shipping" over "discovering value":

* **Engineering as a Service (The "IT" Cost-Center Trap):**

  * **Signal:** The bank consistently separates "Business" from "IT" functions, framing technical departments as service providers for bank divisions rather than core value creators.

  * **Evidence:** "Internal control systems... and approval by the Management Board" are emphasized for digital iterations, indicating a subservient IT relationship.

  * **Risk:** This creates a mercenary culture where engineers are told what to build (output) rather than being responsible for how it solves a customer problem (outcome).

* **Stakeholder-Driven Roadmap (Lack of Empowerment):**

  * **Signal:** High-level strategic planning and operational control are centralized, leaving little room for team-level autonomy.

  * **Evidence:** "The Bank currently operates a multi-level ESG governance system that includes strategic planning [and] operational control." (ESG Report, Page 6)

  * **Risk:** This centralized control prevents Product Discovery. When the Board approves features, teams lose the autonomy to pivot based on user data, resulting in products that meet "business requirements" but fail to delight users.

* **Project-Based Funding vs. Product Persistence:**

  * **Signal:** Strategic financial planning is tied to fixed-term "projects" rather than the continuous funding of durable product teams.

  * **Evidence:** "The Board of Directors approved a new ESG strategy for 2025â€“2027, aimed at integrating ESG factors into business processes." (ESG Report, Page 6)

  * **Evidence:** Financial strategy highlights the "Implementation of IT-projects for 2024," implying fixed-term budgets and finite scopes.

  * **Risk:** A project mindset encourages "mercenary" behavior to meet deadlines rather than building long-term architectural integrity or sustainable customer value.

### **3. Strategic Disconnects**

* **Innovation vs. Maintenance:** While the bank aims for "Ecosystem Leadership," approximately 70% of the IT budget is allocated to maintenance and regulatory compliance, leaving insufficient capital for true disruptive innovation.

* **The "Digital Veneer" Problem:** High digital volumes (92% of loans issued online) represent the digitization of old processes, not necessarily the innovation of new products.

* **Explainability Fatigue:** Internal teams, specifically in Legal and Compliance, are drowning in manual work because current "IT solutions" prioritize speed of delivery over the explainable decision support required by banking professionals.

### **4. The Prescription**

To transition from a "Digital Immigrant" to an Empowered Product Organization, the Board must mandate the following shifts:

* **Abolish the "IT Department" Concept:** Reorganize into Durable Product Teams (e.g., SME Lending, Halyk Travel, GovTech) where engineers, designers, and product managers sit together permanently.

* **Mandate Product Discovery:** Before any "IT project" is funded, teams must prove they have validated the Value, Usability, Feasibility, and Viability of the solution through prototyping and user testing.

* **Shift from "Approving Features" to "Assigning Problems":** The Board should move to an OKR (Objectives and Key Results) model. Instead of demanding "Feature X," the Board should demand an outcome, such as "Reduce Corporate Onboarding Time from 48 hours to 4 hours".

* **Invest in Explainable AI (XAI) for Governance:** Deploy internal agents like "Halyk LegalEye" to bridge the gap between digital speed and legal accountability, ensuring that technology supports human judgment rather than replacing it with "black box" logic.
