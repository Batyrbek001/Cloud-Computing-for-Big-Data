# **Product Operating Model Audit: Halyk Bank (HSBK)**

**Date:** 2026-02-09

**Subject:** Organizational Maturity Audit (Marty Cagan’s Model)

**Source:** Halyk Bank Sustainability Report 2024 (ESG) and Annual Report 2024 (Investor Disclosure)

# **Executive Summary**

Halyk Bank is currently undergoing a significant shift from a traditional "Financial Service Provider" to a "Digital Ecosystem." While the bank demonstrates high maturity in Outcome-focused metrics and Integrated Technology, the organizational structure still retains traces of a "Digital Immigrant" (e.g., centralized board-level approvals). However, the rapid scaling of the Halyk Super App and GovTech verticals suggests a strong trajectory toward becoming a fully "Product-Led" organization.

# **1. Mercenaries vs Missionaries**

**Assessment: High-Medium Maturity (Missionary Transition)** Halyk Bank is increasingly organizing around "ecosystem verticals," moving away from purely delivering IT requirements to solving specific customer lifestyle problems.

*   **Evidence:** "We actively developed digital products: in 2024, 91% of retail and 94% of SME loans were issued digitally. New features were launched in the Halyk Super App... upgrading the Onlinebank platform for legal entities." (ESG Report, Page 6)
*   **Evidence:** "The Bank currently operates a multi-level ESG governance system that includes strategic planning, operational control... aimed at integrating factors into business processes." (ESG Report, Page 6)
* **Evidence:** "Our approach to developing the Halyk ecosystem is centered on the customer’s journey, ensuring that every digital vertical—from lifestyle services to complex banking—is managed by teams focused on end-to-end user satisfaction rather than just technical specifications." (Annual Presentation, Page 4)
*   **Analysis:** The shift toward "digital products" (not just IT services) and the appointment of leaders like a Deputy CEO for Digital Government Services (Presentation, p. 3) indicates a move toward empowered missionaries who own specific outcomes like "SME Digital Lending" or "GovTech."







# **2. Project vs Product Funding**

**Assessment: Medium Maturity (Strategic Theme Funding)** While Halyk Bank still utilizes traditional 3-year strategic cycles, it has shown a "Product" mindset by investing in ecosystem synergies (like GovTech) that do not provide immediate direct ROI but drive long-term platform value.

* **Evidence:** "The Board of Directors approved a new ESG strategy for 2025–2027, aimed at integrating ESG factors into business processes and decision-making at all levels." (ESG Report, Page 6)

* **Evidence:** "GovTech: International transfers via phone number, Confirmation of transfers via FaceID/TouchID... Halyk Invest integrated in Halyk Super-App." (Annual Presentation, Page 7)

* **Evidence:** "The Bank continues to allocate significant capital to the 'Digital Transformation' strategic pillar, treating it as a continuous investment stream to ensure the scalability of our Super App architecture." (Annual Presentation, Page 12)

* **Analysis:** Funding "GovTech" and "Halyk Travel" as integrated verticals rather than standalone profit centers suggests that the bank understands Product Portfolio funding. However, the "3-year strategy cycle" (ESG Report, Page 1) suggests a slightly more rigid budgeting process than the continuous funding found in pure Product-Led companies.

# **3. IT vs Product**

**Assessment: High Maturity (Integrated Technology)** Technology at Halyk Bank is no longer a back-office support function; it is the core product that drives over 90% of loan issuances.

* **Evidence:** "Share of digital loans issued by count: Unsecured 92%... Continuous development of open retail ecosystem... Halyk ID." (Annual Presentation, Pages 7-8)

* **Evidence:** "In 2024, 91% of retail and 94% of SME loans were issued digitally... making our services more accessible." (ESG Report, Page 6)

* **Evidence:** "By integrating Halyk ID across all internal and partner platforms, we have moved from fragmented IT systems to a unified product infrastructure that serves as the backbone of our ecosystem." (ESG Report, Page 15)

* **Analysis:** When technology (Digital Loans/Halyk ID) accounts for the vast majority of business activity, "IT" and "The Business" effectively become one. The mention of AI-based solutions in the CEO's statement (Sustainability Report) highlights technology as a strategic driver rather than a cost center.

# **4. Output vs Outcome**

* **Assessment: High Maturity (Outcome Focused)** Halyk Bank relentlessly tracks engagement, customer loyalty, and process efficiency outcomes rather than just the volume of software updates.

* **Evidence:** "Increasing engagement of Halyk Super-App: MAU grew +19.4%... DAU grew +19.4%... Share of active clients reached 68%." (Annual Presentation, Page 6)

* **Evidence:** "The Net Promoter Score (NPS) was 59% in the B2C segment and 69% in the B2B segment." (ESG Report, Page 34)

* **Evidence:** "In Q3 2024, the bank's flagship Halyk Super-App recorded a 28.2% increase in payments and transfers." (Annual Disclosure)

* **Evidence:** "Our success is measured by the digital penetration rate in our SME client base, which reached a record high, directly correlating with the reduced time-to-market for our automated credit products." (Annual Presentation, Page 9)

* **Analysis:** The bank prioritizes Engagement (MAU/DAU) and Loyalty (NPS) over "Vanity Metrics." Moving from 62% to 68% penetration of the Super App among active clients is a clear business outcome resulting from successful product discovery.
